```python
%matplotlib inline
from sklearn import datasets
from sklearn import tree
import pandas as pd
import numpy as np
import seaborn as sns


iris = datasets.load_iris()

df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])

df.head()
```

```python
df.describe()
```

```python
sns.set_style('whitegrid')
sns.pairplot(df, hue='target')
```

```python
first_sample = 1; last_sample = 149
X = iris.data[first_sample:last_sample, :]
X.shape
```

```python
Y = iris.target[first_sample:last_sample]
Y.shape
```

```python
test_setosa = [iris.data[0, :]]
test_setosa, iris.target_names[iris.target[0]]
```

```python
df.tail()
```

```python
test_virginica = [iris.data[149, :]]
test_virginica, iris.target_names[iris.target[149]]
```

```python
# Decision Tree Classifier Model
model_tree = tree.DecisionTreeClassifier()

# Training the model
model_tree.fit(X, Y)

pred_tree_setosa = model_tree.predict(test_setosa)
print('Decision Tree predicts {} for test_setosa'.format(iris.target_names[pred_tree_setosa]))
```

```python
pred_tree_virginica = model_tree.predict(test_virginica)
print('Decision Tree predicts {} for test_virginica'.format(iris.target_names[pred_tree_virginica]))
```
